```python
# TASK 3: Calculate the distances between a launch site to its proximities

import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    # convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    R = 6371  # Earth radius in km
    return R * c

# Apply it to each site
launch_sites_df['Distance_to_JSC_km'] = launch_sites_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], 29.559684888503615, -95.0830971930759),
    axis=1
)

# Inspect
print(launch_sites_df[['Launch Site', 'Distance_to_JSC_km']])
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
import numpy as np

# 1) Define a Haversine function
def haversine(lat1, lon1, lat2, lon2):
    # Convert from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2.0)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    R = 6371  # Earth radius in km
    return R * c

# 2) Reference points
nasa_jsc = (29.559684888503615, -95.0830971930759)  # NASA JSC
coastline_ref = (28.56367, -80.57163)                # Example coastline near CCAFS

# 3) Compute distances for each launch site
launch_sites_df['Distance_to_JSC_km'] = launch_sites_df.apply(
    lambda r: haversine(r['Lat'], r['Long'], nasa_jsc[0], nasa_jsc[1]), axis=1)

launch_sites_df['Distance_to_Coast_km'] = launch_sites_df.apply(
    lambda r: haversine(r['Lat'], r['Long'], coastline_ref[0], coastline_ref[1]), axis=1)

# 4) Inspect results
print(launch_sites_df[['Launch Site', 'Distance_to_JSC_km', 'Distance_to_Coast_km']])
```
